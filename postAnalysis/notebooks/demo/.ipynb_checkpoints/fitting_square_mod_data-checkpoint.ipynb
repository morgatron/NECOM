# Find an appropriate window to analyse
We want a full number of periods for both modulations

```python
# Find the longest section where both modulations have integer numbers of periods
modLabels = ["Bz_1", "Bx_1"]
modulation_periodsD = {label:ds.params.modsSynced[label].period_cycles for label in modLabels}
lcm = math.lcm(*modulation_periodsD.values()) # The lowest common multiple of these periods
n_lcms = int(ds.T.size/lcm-0.5) #The number of LCMs that can safely be fit given the number of traces taken
slice_len = n_lcms*lcm # This is the length of the section we'll take
print(slice_len)
```

```python
# Now find the index to start the analysis. We'll look for the first time where the periods are in phase, i.e. both(all) rise together.
flag_idxs = (6,7) #Bz, Bx
mag_flags = [ds.flags & 2**idx for idx in flag_idxs]
start_idx = au.index_of_coinciding_first_rising_edges(mag_flags)
stop_idx = slice_len+start_idx
signalA = ds.signal[start_idx:stop_idx]
signalA
```

# Split it up and calculate the mean responses

```python
def calculate_mean_response(signals, period):

    # Reshape into
    response_ = signals.reshape(-1,period, ds.t.size).mean(axis=0)
    rp, rm = response_[:period//2], response_[-(period//2):]
    full_response = ((rp-rm[-1]) - (rm-rp[-1]))[1:]

    #fast response, i.e. that immediately after a field is turned on
    magnetic_signature = full_response[0:3].mean(axis=0)  
    N = magnetic_signature.size//2
    magnetic_signature = (magnetic_signature[:N])# + magnetic_signature[-N:])/2
    return full_response, magnetic_signature

# Convert to numpy array so we can use reshape etc.
sig = signalA.values.astype('i4')

resD = {key: dict(zip(["response", "halfSig"], calculate_mean_response(sig, modulation_periodsD[key])))
                 for  key in ['Bx_1', 'Bz_1']}
#resD = {key:dict(response=response, halfSig = halfSig) for key in ['Bxh}
del sig
```

# Fitting time

The "response" arrays give the 'signatures' of a step perturbation as a function of time since the perturbation started. 

We now fit these 'signatures' to get the fields seen by the alkali in each case.

The very first point after the perturbation may be interpreted as a response of the alkali alone to the magnetic field step. We use this to determine what an "x" or "z" field seen by the alkali looks like. We then fit the later signatures using these traces to get the field seen by the alkali as a function of time since the step change.
